```python
min = -0.5
max = 0.5
size = 1000000

uniform_data = np.random.uniform(min, max, size)

# 绘制直方图
plt.figure(figsize=(8, 5))
plt.hist(uniform_data, bins=100, color='skyblue')
plt.title("Weight Uniform Distribution (low=-0.5, high=0.5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
weight_entropy = []
total = len(uniform_data)

for i in range(len(uniform_data)):
  percent = (i / float(total)) * 100
  print(f"\r计算进度: {percent:.4f}%", end="")

  w = abs(uniform_data[i]) / sum(abs(uniform_data))
  weight_entropy.append(-1 * w * np.log(w))

print("\n计算完成！")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function
x = np.linspace(0.0001, 500, 5000)
y1 = -x * np.log(x)
y2 = -(x / 10) * np.log(x / 10)
y3 = -(x / 50) * np.log(x / 50)
y4 = -(x / 100) * np.log(x / 100)
y5 = -(x / 500) * np.log(x / 500)
y6 = -(x / 1000) * np.log(x / 1000)

# Plot
plt.plot(x, y1, label=r"$y = -x \ln(x)$", color="blue")
plt.plot(x, y2, label=r"$y = -\frac{x}{10} \ln\left(\frac{x}{10}\right)$", color="orange")
plt.plot(x, y3, label=r"$y = -\frac{x}{50} \ln\left(\frac{x}{50}\right)$", color="green")
plt.plot(x, y4, label=r"$y = -\frac{x}{100} \ln\left(\frac{x}{100}\right)$", color="purple")
plt.plot(x, y5, label=r"$y = -\frac{x}{500} \ln\left(\frac{x}{500}\right)$", color="brown")
plt.plot(x, y6, label=r"$y = -\frac{x}{1000} \ln\left(\frac{x}{1000}\right)$", color="pink")

# Adding horizontal y=0.36788 and reference lines y=0, x=0
plt.axhline(0.36788, color='red', linewidth=1.2, linestyle='--', label="y = 0.36788")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # reference lines y=0
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # reference lines x=0

# Setting
plt.ylim(-1, 1)  # Range of y
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of Functions with Y-axis (-1, 1)")
plt.grid()
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 1))

# Display
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数和变量范围
x = np.linspace(0.0001, 5, 500)  # 避免 x 为 0 导致分母为 0
y = np.linspace(1, 10, 5)
X, Y = np.meshgrid(x, y)

# 计算函数值
Z = -(X / Y) * np.log(X / Y)

# 限制 z 的范围在 [-1, 1]
Z = np.clip(Z, -1, 1)

# 绘制图像
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r"$z = -\frac{x}{y}\ln\left(\frac{x}{y}\right)$")

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
plt.show()
```
